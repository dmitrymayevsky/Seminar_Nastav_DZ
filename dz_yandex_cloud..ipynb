```scala
%spark.pyspark
from pyspark.sql import functions as F
from pyspark.sql.window import Window

users = spark.createDataFrame([
    ("u1", "Berlin"),
    ("u2", "Berlin"),
    ("u3", "Munich"),
    ("u4", "Hamburg")
], ["user_id", "city"])

orders = spark.createDataFrame([
    ("o1", "u1", "p1", 2, 10.0),
    ("o2", "u1", "p2", 1, 30.0),
    ("o3", "u2", "p1", 1, 10.0),
    ("o4", "u2", "p3", 5, 7.0),
    ("o5", "u3", "p2", 3, 30.0),
    ("o6", "u3", "p3", 1, 7.0),
    ("o7", "u4", "p1", 10, 10.0)
], ["order_id", "user_id", "product_id", "qty", "price"])

products = spark.createDataFrame([
    ("p1", "Ring VOLA"),
    ("p2", "Ring POROG"),
    ("p3", "Ring TISHINA")
], ["product_id", "product_name"])

print("Успех")
users.show()
orders.show()
products.show()
```

```scala
%spark.pyspark

orders_with_revenue = orders.withColumn("revenue", F.col("qty") * F.col("price"))

joined_df = orders_with_revenue \
    .join(users, "user_id", "left") \
    .join(products, "product_id", "left")
    
aggregated_df = joined_df.groupBy(
    "city", "product_id", "product_name"
).agg(
    F.count("order_id").alias("orders_cnt"),
    F.sum("qty").alias("qty_sum"),
    F.sum("revenue").alias("revenue_sum")
)

print("Выполнена агрегация")
aggregated_df.show()
```

```scala
%spark.pyspark

window_spec = Window.partitionBy("city").orderBy(F.desc("revenue_sum"))

top_products_df = aggregated_df \
    .withColumn("rank", F.row_number().over(window_spec)) \
    .filter(F.col("rank") <= 2) \
    .drop("rank")
    
print("Тop-2 товара")
top_products_df.orderBy("city", F.desc("revenue_sum")).show()
```

```scala
%spark.pyspark


BUCKET_NAME = "spark-bucket-homework" 

hdfs_path = "/tmp/sandbox_zeppelin/mart_city_top_products/"
s3_path = f"s3a://{BUCKET_NAME}/mart_city_top_products/"

top_products_df.write.mode("overwrite").parquet(hdfs_path)
print(f"HDFS: {hdfs_path}")

top_products_df.write.mode("overwrite").parquet(s3_path)
print(f"S3: {s3_path}")
```
